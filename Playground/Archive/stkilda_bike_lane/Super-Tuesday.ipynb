```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "C:/Users/Minh Pro/Downloads/annual-bike-counts-super-tuesday.csv" 
bike_count_df = pd.read_csv(file_path)

# Filter data for locations near St Kilda Road
stkilda_mask = bike_count_df['description'].str.contains('St Kilda Rd', case=False, na=False)
stkilda_df = bike_count_df[stkilda_mask]

# Group the data by year and calculate the average count for each year
yearly_average_counts = stkilda_df.groupby('year')['total'].mean().reset_index()

# Plot the average bike count over the years
plt.figure(figsize=(10, 5))
plt.plot(yearly_average_counts['year'], yearly_average_counts['total'])
plt.xlabel('Year')
plt.ylabel('Average Bike Count')
plt.title('Average Bike Count Near St Kilda Road (Annual Super Tuesday Event)')
plt.xticks(yearly_average_counts['year'], rotation=45)
plt.show()

# Print the DataFrame to the console
print(yearly_average_counts)
```

```python
# Import necessary library
import pandas as pd

# Load the dataset
file_path = "C:/Users/Minh Pro/Downloads/annual-bike-counts-super-tuesday.csv"
bike_df = pd.read_csv(file_path)

# Create a smaller DataFrame with just the time slot columns
times_df = bike_df[['7:00 AM', '7:15 AM', '7:30 AM', '7:45 AM', '8:00 AM', '8:15 AM', '8:30 AM', '8:45 AM']]

# Sum the counts in each time slot across all days
time_totals = times_df.sum()

# Find the time slot with the maximum total count
peak_time = time_totals.idxmax()

print(f"The peak time slot for bike usage is: {peak_time}")
```

```python
# Plot the sums in a bar chart
plt.figure(figsize=(10, 6))
time_totals.plot(kind='bar', color='blue')
plt.title('Total Bike Counts for Each Time Slot')
plt.xlabel('Time Slot')
plt.ylabel('Total Bike Counts')
plt.show()
```

The peak bike usage at 8:30 AM on the St Kilda Rd bike lane might be due to several factors, primarily associated with commuting patterns:

- **Commute to Work/School**: 8:30 AM is a common start time for work and school. The peak at this time suggests that many people are using the bike lane for their morning commute.

- **Exercise**: Some people might be using this time for morning exercise before starting their work or school day.

- **Traffic Avoidance**: Cycling can be a quicker and more predictable mode of transportation during peak traffic hours. People might choose to bike to avoid traffic congestion at this time.

- **Environmental Factors**: If the weather is typically cool and pleasant in the morning, more people might choose to cycle at this time.

- **Cycling Infrastructure**: If St Kilda Rd is well-equipped with safe and efficient cycling paths, it could contribute to the high usage at this peak time.
