```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Replace 'your_download_link_here' with the actual download link
download_link = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/residential-dwellings/exports/csv?lang=en&timezone=Australia%2FSydney&use_labels=true&delimiter=%2C'

# Read the CSV into a DataFrame
bay_df = pd.read_csv(download_link)

bay_df
```

```python
# 假设 df 中有 'latitude' 和 'longitude' 列
gdf = gpd.GeoDataFrame(
    bay_df, 
    geometry=gpd.points_from_xy(bay_df.Longitude, bay_df.Latitude),
    crs="EPSG:4326"  # 设置坐标参考系统，这里假设是WGS84
)
```
