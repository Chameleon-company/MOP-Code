# **Import modules & fetch data from Open Data Melbourne**

```python
import requests
import pandas as pd
import os


# define function to fetch data from website using API

def fetch_data(base_url, dataset, api_key, num_records = 99, offset = 0):
    all_records = []
    max_offset = 9900  # Maximum number of requests

    while True:
        # maximum limit check
        if offset > max_offset:
            break

        # Create API request URL
        filters = f'{dataset}/records?limit={num_records}&offset={offset}'
        url = f'{base_url}{filters}&api_key={api_key}'

        # Start request
        try:
            result = requests.get(url, timeout=10)
            result.raise_for_status()
            records = result.json().get('results')
        except requests.exceptions.RequestException as e:
            raise Exception(f"API request failed: {e}")
        if records is None:
            break
        all_records.extend(records)
        if len(records) < num_records:
            break

        # next cycle offset
        offset += num_records

    # DataFrame all data
    df = pd.DataFrame(all_records)
    return df

API_KEY = os.environ.get('MELBOURNE_API_KEY', input("Please enter your API key: "))
BASE_URL = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
```

```python
# initialize a figure with Plotly
fig_bub_map = go.Figure()

# Define the geographic center as a dictionary
map_center = dict(lat=geo_total.iloc[0]['geo_point_2d']['lat'], lon=geo_total.iloc[0]['geo_point_2d']['lon'])

# scaling factor for bubble sizes
scale_factor = 0.1

# loop through each district to add bubbles
for index, row in geo_total.iterrows():

    # calculate the center of each district if not already present in geo_point_2d
    center_lat, center_lon = row['geo_point_2d']['lat'], row['geo_point_2d']['lon']

    # calculate the proportions of each development type
    total_developments = row['total']

    # skip if total developments is zero
    if total_developments == 0:
      continue

    # calculate proportions and scale bubble size
    bubble_size = total_developments * scale_factor

    # adding bubble for CLUE areas
    fig_bub_map.add_trace(go.Scattergeo(
        #locationmode = 'ISO-3',
        lon = [center_lon],
        lat = [center_lat],
        text = f"{row['featurenam']}: {total_developments} Developments",
        marker = dict(
            size = bubble_size,  # Size of the bubble
            color = 'blue'
        )
    ))


# Set up the layout for the map
fig_bub_map.update_layout(
    title = 'Interactive Bubble Map of Developments',
    geo = dict(
        showland = True,
        landcolor = "rgb(212, 212, 212)",
        subunitcolor = "rgb(255, 255, 255)",
        countrycolor = "rgb(255, 255, 255)",
        showlakes = True,
        lakecolor = "rgb(255, 255, 255)",
        center = map_center,
        projection_scale = 10,    # to be tuned to appropriately scale the map to the data
    )
)

# Show the figure
fig_bub_map.show()
```
