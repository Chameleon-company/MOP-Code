######**Summary:** City events and works permit information that have date and location information.

######*This dataset contains active and planned activities across the City of Melbourne area. Activities include city works (such as proposed road closures, footpath works or reserved parking permits) and event bookings. The information is derived from permits which are issued by City of Melbourne.*

######Note: *Each activity includes geo-spatial information to indicate the general area of the activity. The spatial information is not indicative of the activity size as the activity may only take up a small portion of the area for the entry.*

######Note: *The start and end date fields indicate that the event/activity is expected to occur within this time range. The actual activity may be active for significantly shorter duration than the start and end dates.*


######*Note this API has been retired*

```python
def fetch_data(base_url, dataset, api_key, num_records=99, offset=0):
    all_records = []
    max_offset = 9900

    while True:
        if offset > max_offset:
            break

        filters = f'{dataset}/records?limit={num_records}&offset={offset}'
        url = f'{base_url}{filters}&api_key={api_key}'

        try:
            result = requests.get(url, timeout = 10)
            result.raise_for_status()
            records = result.json().get('results')
        except requests.exceptions.RequestException as e:
            raise Exception(f'API request failed: {e}')
        if records is None:
            break
        all_records.extend(records)
        if len(records) < num_records:
            break

        offset += num_records

    df = pd.DataFrame(all_records)
    return df
```

```python
#Median activity count for each year

# Filter activities_encode include only attributes for start year and month
activities_date = activities_encode[['year_start', 'month_start']]

# Add a new column 'activity_count' with a value of 1 for each row (day)
activities_date['activity_count'] = 1

# Group year_start & month_start to calculate the count of development activities in each month
activities_count = activities_date.groupby(['year_start', 'month_start']).agg({'activity_count': 'sum'}).reset_index()

# Create pivot table to reshape the data & have years as rows & months as columns
activities_pivot = activities_count.pivot(index='year_start', columns='month_start', values='activity_count')

# Calculate median activity count for each year
median_activity_per_year = activities_pivot.median(axis=1)

# Plot line chart for median activity count per year
median_activity_per_year.plot(kind='line', figsize=(10, 6), marker='o', linestyle='-', color='b')
plt.title('Median Activity Count per Year')
plt.xlabel('Year')
plt.ylabel('Median Activity Count')
plt.show()
```

```python
#Filter activities_encode
#Columns related to dates
activities_seasonal = activities_encode[['year_start', 'month_start']]

#Map months to Melbourne seasons
def melbourne_season(month):
    if month in [12, 1, 2]:
        return 'Summer'
    elif month in [3, 4, 5]:
        return 'Autumn'
    elif month in [6, 7, 8]:
        return 'Winter'
    elif month in [9, 10, 11]:
        return 'Spring'
    else:
        return 'Unknown'

#New column for function
activities_seasonal['melbourne_season'] = activities_seasonal['month_start'].apply(melbourne_season)

#New column with value of 1 for each row
activities_seasonal['activity_count'] = 1

#Group to calculate the count of development activities in seasons
activities_count = activities_seasonal.groupby(['year_start', 'melbourne_season']).agg({'activity_count': 'sum'}).reset_index()

#Pivot table to reshape the data
activities_pivot = activities_count.pivot(index='year_start', columns='melbourne_season', values='activity_count')

#Plot
# activities_pivot.plot(kind='box', figsize=(10, 6)) # Bot plot
activities_pivot.plot(kind='line', marker='o', figsize=(10, 6)) # Line chart
plt.title('Seasonal Trends in Development Activities (Melbourne)')
plt.xlabel('Year')
plt.ylabel('Activity Count')
plt.legend(title='Melbourne Season', bbox_to_anchor=(1, 1))
plt.show()
```
