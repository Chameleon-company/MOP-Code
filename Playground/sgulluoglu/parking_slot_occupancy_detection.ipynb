(matching with the other column-row values) and save it to a new csv file and put it into repository. 

```python
import requests
import pandas as pd

# API link (example)
api_url = "https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/on-street-parking-bay-sensors/records?limit=100&offset=6000"  

# Fetch the data from the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    # Convert the data to a DataFrame
    df = pd.DataFrame(data)
    
    # Select 99 random rows of the DataFrame
    random_sample = df.sample(n=99)
    print("Random 99 rows of the DataFrame:")
    print(random_sample)
    
    # Identify missing values
    missing_values = random_sample.isnull()
    print("\nMissing Values (True indicates missing value):")
    print(missing_values)
    
    # Count missing values in each column
    missing_count = random_sample.isnull().sum()
    print("\nCount of Missing Values in Each Column:")
    print(missing_count)
    
    # List related rows or columns with missing values
    print("\nRows with Missing Values:")
    rows_with_missing_values = random_sample[random_sample.isnull().any(axis=1)]
    print(rows_with_missing_values)

    # Fill missing values by matching with other columns
    for column in df.columns:
        if df[column].isnull().any():
            if df[column].dtype == 'object':
                # For categorical columns, fill with the most frequent value (mode) in the corresponding rows
                most_frequent = df.groupby('another_column')[column].apply(lambda x: x.mode()[0] if not x.mode().empty else None)
                df[column] = df.apply(lambda row: most_frequent[row['another_column']] if pd.isnull(row[column]) else row[column], axis=1)
            else:
                # For numerical columns, fill with the median value in the corresponding rows
                median_value = df.groupby('another_column')[column].median()
                df[column] = df.apply(lambda row: median_value[row['another_column']] if pd.isnull(row[column]) else row[column], axis=1)
    
    print("\nDataFrame after filling missing values:")
    print(df.head(99))

    # Verify that there are no missing values left
    missing_count_after = df.isnull().sum()
    print("\nCount of Missing Values in Each Column after filling:")
    print(missing_count_after)
    
    # Save the cleaned DataFrame to a new CSV file
    output_file_path = 'cleaned_data.csv'
    df.to_csv(output_file_path, index=False)
    print(f"\nCleaned DataFrame saved to {output_file_path}")
else:
    print(f"Failed to fetch data from the API. Status code: {response.status_code}")
```
