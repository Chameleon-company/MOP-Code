```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

print("pandas:", pd.__version__)
print("geopandas:", gpd.__version__)
print("Setup complete!")
```

Dataset 1: Urban Forest Tree Species 

```python
#Load Data (Urban Forest Tree Dataset)
trees = gpd.read_file("../data/raw/trees-with-species-and-dimensions-urban-forest.geojson")

# Basic shape and info
print(f"Shape: {trees.shape}")
print(f"\nColumn names:\n{trees.columns.tolist()}")
print(f"\nData types:\n{trees.dtypes}")
print(f"\nFirst 5 rows:")
trees.head()
```

```python
# Missing values
print(f"Missing values:\n{trees.isnull().sum()}")
print(f"\nMissing value percentages:\n{(trees.isnull().sum() / len(trees) * 100).round(1)}")
```

```python
# Key categorical distributions
print("Useful Life Expectancy:")
print(trees['useful_life_expectency'].value_counts())
print(f"\nAge Description:")
print(trees['age_description'].value_counts())
```

```python
print("Useful Life Expectancy Value:")
print(trees['useful_life_expectency_value'].describe())
print(f"\nValue counts:\n{trees['useful_life_expectency_value'].value_counts().sort_index()}")
```

```python
# DBH distribution (for non-missing values)
print("Diameter Breast Height (non-null):")
print(trees['diameter_breast_height'].describe())

print(f"\nTop 10 species:")
print(trees['common_name'].value_counts().head(10))
```

```python
# Quick spatial check
print(f"Latitude range: {trees['latitude'].min():.4f} to {trees['latitude'].max():.4f}")
print(f"Longitude range: {trees['longitude'].min():.4f} to {trees['longitude'].max():.4f}")
print(f"\nPrecinct distribution:")
print(trees['precinct'].value_counts())
```

Dataset 2: Microclimate Sensor Data

```python
# Load microclimate sensor data
sensors = pd.read_csv("../data/raw/microclimate-sensors-data.csv")

print(f"Shape: {sensors.shape}")
print(f"\nColumn names:\n{sensors.columns.tolist()}")
print(f"\nData types:\n{sensors.dtypes}")
print(f"\nFirst 5 rows:")
sensors.head()
```

```python
# Missing values for key columns
print("Missing values:")
print(sensors[['AirTemperature', 'RelativeHumidity', 'LatLong', 'Time']].isnull().sum())

# Temperature and humidity ranges
print(f"\nAir Temperature:")
print(sensors['AirTemperature'].describe())
print(f"\nRelative Humidity:")
print(sensors['RelativeHumidity'].describe())

# How many unique sensors
print(f"\nUnique sensors: {sensors['Device_id'].nunique()}")
print(f"\nSensor locations:")
print(sensors['SensorLocation'].value_counts())
```
