## Netherlands pipe failure datasets (Anne)

This section reviews the Netherlands dataset to check whether it is suitable for our project.

Two files are available:
- a water mains profile dataset in `.pickle` format
- a break history dataset in `.npy` format

The main aim here is to understand the structure of both files and assess how easy they would be to use for our analysis.

```python
#Import Netherlands dataset

#Load pickle file (water mains profile dataset)
mains_nether = pd.read_pickle("../data/raw/Netherlands/Netherlands_watermains.pickle")
mains_nether = mains_nether.drop(columns='unit_ID')

#Load npy file (break history dataset)
breaks_nether = np.load("../data/raw/Netherlands/Netherlands_watermainbreaks.npy")

#Check
print(mains_nether.head())
print(breaks_nether.shape)
```

The break history file is not in a normal 2D table format. Its shape is (10203, 6, 2), meaning it is a 3D array.

To inspect it more easily, it is reshaped into 2D first.

```python
#npy is a 3D array, not a 2D table
#So need to reshape it into 2D first

#Reshape from (10203, 6, 2) to (10203, 12)
breaks_nether = breaks_nether.reshape(breaks_nether.shape[0], -1)

#Convert Numpy array to Dataframe
breaks_nether_df = pd.DataFrame(breaks_nether)

#Check
print(breaks_nether_df.head())
print(breaks_nether_df.shape)
print(breaks_nether_df.info())
```

### Review

**Pros**

- Clean and already prepared for modelling
- Has both profile features + historical failure sequence
- Include some environment-related features (soil/vegetation)

**Cons**

- The break history data is not in a simple tabular format
- Look like it's built more for RNN / sequence model, not ideal for RF/XGBoost
- Need extra work to convert sequence data into flat features
- Unclear schema with no clear column names
- Not public from open government/organisation data sources
