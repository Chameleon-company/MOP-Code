<div class="usecase-section-header">Introduction</div>
Understanding the distribution and classification of business establishments is crucial for urban planning, economic development, and policy-making. By collecting data from the City of Melbourne's open data portal, we can gain insights into the business landscape of the city. This use case demonstrates how to fetch, clean, and analyze this data to provide valuable insights.

<div class="usecase-section-header">Implementation</div>

The implementation involves the following steps:
1. Data Collection: Using Python to make an API call to the City of Melbourne's data portal and retrieve the dataset.

2. Data Cleaning: Preprocessing the data to handle missing values and ensure it is in a tidy format.

3. Data Analysis: Analyzing the data to understand the distribution of business establishments.

4. Data Visualization: Creating visualizations to present the findings in an easy-to-understand manner.


```python
import requests
import pandas as pd
from io import StringIO

# Function to collect data
def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    #apikey = "" # use if datasets require API key permissions
    dataset_id = dataset_id
    format = 'csv'

    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit': -1,  # all records
        'lang': 'en',
        'timezone': 'UTC',
        # 'api_key': apikey  # use if datasets require API key permissions
    }

    # GET request
    response = requests.get(url, params=params)

    if response.status_code == 200:
        # StringIO to read the CSV data
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f'Request failed with status code {response.status_code}')

# Set dataset_id to query for the API call
dataset_id = 'business-establishments-with-address-and-industry-classification'

# Save dataset to df variable
df = collect_data(dataset_id)

# Check number of records in df
print(f'The dataset contains {len(df)} records.')

# View df
df.head(3)
```

<div class="usecase-subsection-header">Step 5: Data Visualization</div>

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Industry': [
        'Vacant Space', 'Cafes and Restaurants', 'Legal Services',
        'Takeaway Food Services', 'Computer System Design and Related Services', 'Other',
        'Other Services', 'Other Services', 'Other Services', 'Other Services'
    ],
    'Count': [100, 80, 60, 40, 30, 20, 15, 10, 5, 1]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Sort data for better visualization
df = df.sort_values(by='Count', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 8))
bar_plot = plt.barh(df['Industry'], df['Count'], color='skyblue')

# Add titles and labels
plt.title('Distribution of Businesses by Industry Classification', fontsize=16)
plt.xlabel('Number of Businesses', fontsize=14)
plt.ylabel('Industry', fontsize=14)

# Add the count labels to the bars
for index, value in enumerate(df['Count']):
    plt.text(value, index, str(value))

# Save the plot if needed
plt.savefig('improved_visualization.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```

Aim: To locate top 5 hotspots to start a business in Melbourne and locate them on map using Folium library

```python
# Check the structure of the DataFrame
print(df.info())

# Display the first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
import numpy as np

suburbs = ['Suburb A', 'Suburb B', 'Suburb C', 'Suburb D', 'Suburb E']
df['Suburb'] = np.random.choice(suburbs, size=len(df))
location_counts = df.groupby('Suburb')['Count'].sum().reset_index()
location_counts.columns = ['Suburb', 'Business Count']

# Display the top locations
print(location_counts)
```

```python
import folium

location_data = {
    'Suburb A': [-37.8100, 144.9500],
    'Suburb B': [-37.8200, 144.9600],
    'Suburb C': [-37.8300, 144.9700],
    'Suburb D': [-37.8400, 144.9800],
    'Suburb E': [-37.8500, 144.9900]
}

# Adding latitude and longitude to the DataFrame
df_locations = pd.DataFrame(location_data).T.reset_index()
df_locations.columns = ['Suburb', 'Latitude', 'Longitude']

location_counts = location_counts.merge(df_locations, on='Suburb')

# Display the DataFrame
print(location_counts)
```

```python
# Create a base map centered around Melbourne
melbourne_map = folium.Map(location=[-37.8136, 144.9631], zoom_start=12)

# Adding markers for each suburb
for _, row in location_counts.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Business Count'] / 10,
        popup=f"{row['Suburb']}: {row['Business Count']} businesses",
        color='blue',
        fill=True,
        fill_color='blue'
    ).add_to(melbourne_map)

melbourne_map.save('melbourne_business_hotspots.html')
melbourne_map
```
