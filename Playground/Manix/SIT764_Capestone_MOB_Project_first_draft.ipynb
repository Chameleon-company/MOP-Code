<h3 style='color:blue;'>Converting Geo Points to Human-Readable Locations</h3>

To make our energy consumption data more relatable and suitable for urban planning, we'll convert the 'Geo Point' coordinates into human-readable locations (e.g., suburb or city names). This process is called **reverse geocoding**, where we use latitude and longitude to find a physical address or area name. We'll use the `geopy` library for this.

First, we need to install the `geopy` library.

```python
!pip install geopy
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="my-geocoder-app")

# Wrap the geocoding function with a rate limiter to avoid overloading the service
# This adds a delay between requests, so it might take some time for large datasets.
geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)

# Function to extract latitude and longitude and then geocode
def get_location_details(geo_point):
    try:
        # Geo Point is like '-37.83363154012436, 144.98420174077557'
        lat, lon = map(float, geo_point.split(','))
        location = geocode((lat, lon), language='en')
        if location and location.address:
            # Extract a suitable human-readable location, e.g., suburb or city
            # You might need to adjust this based on the Nominatim response structure
            address_parts = location.address.split(', ')
            # Attempt to find a city or suburb name, prioritizing clarity
            # This is a heuristic and might need adjustment based on typical results
            for part in address_parts:
                if 'Melbourne' in part or 'VIC' in part or 'Australia' in part:
                    continue # Skip general city/state/country
                if not part.isdigit() and not any(char.isdigit() for char in part):
                    return part # Return the first non-numeric part as a potential suburb/area
            return address_parts[0] # Fallback to the first part if no specific suburb found
        return None
    except Exception as e:
        # print(f"Error geocoding {geo_point}: {e}")
        return None

print("Extracting location details from Geo Points. This may take a while...")

# Apply the function to the 'Geo Point' column to create a new 'Location' column
# Use .progress_apply for a progress bar if tqdm is installed (optional)
# from tqdm.notebook import tqdm
# tqdm.pandas()
# df_energy_processed['Location'] = df_energy_processed['Geo Point'].progress_apply(get_location_details)
df_energy_processed['Location'] = df_energy_processed['Geo Point'].apply(get_location_details)


print("Geocoding complete. Displaying first few rows with new 'Location' column.")
display(df_energy_processed.head())
print(df_energy_processed['Location'].value_counts(dropna=False).head())

# Now, we will recreate df_energy_long using the new 'Location' column in id_vars
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Get the processed dataframe for energy consumption
# df_energy_processed is already updated with 'Location' column from the previous step

print("Columns in df_energy_processed:")
print(df_energy_processed.columns)

# Identify columns that represent energy consumption for different years
year_columns = [col for col in df_energy_processed.columns if 'total_' in col and col.replace('total_','') .isdigit()]

# Reshape the dataframe using `melt` to convert year columns into rows
# Now include 'Location' along with 'Geo Point' as id_vars
df_energy_long = df_energy_processed.melt(id_vars=['Geo Point', 'Location'],
                                          value_vars=year_columns,
                                          var_name='Year_Str',
                                          value_name='Total_Energy_Consumption_kWh')

# Extract the year from the 'Year_Str' column and convert to datetime
df_energy_long['Year'] = df_energy_long['Year_Str'].str.replace('total_', '').astype(int)
df_energy_long['Date'] = pd.to_datetime(df_energy_long['Year'], format='%Y')

# Drop the original 'Year_Str' column as we now have 'Year' and 'Date'
df_energy_long = df_energy_long.drop(columns=['Year_Str'])

print("Reshaped df_energy_consumption_processed into a time-series friendly format, including Location.")
display(df_energy_long.head())
print(df_energy_long.info())
```

<h2 style='color:blue;'>Conceptual Data Structuring for Temporal Fusion Transformers (TFT)</h2>

Temporal Fusion Transformers (TFTs) are a powerful deep learning architecture for multi-horizon time series forecasting. They are designed to handle complex temporal dynamics, including multi-variate inputs, static (time-invariant) covariates, and dynamic (time-varying) covariates. Key requirements for TFTs include:

*   **Time Index:** A clear time variable (e.g., `Date`).
*   **Target Variable:** The variable to be forecasted (e.g., `Value` for population).
*   **Grouping Variables (Entities):** Variables that define individual time series (e.g., `Geography`, `Gender`, `Age`). A TFT can forecast multiple series simultaneously.
*   **Static Covariates:** Features that do not change over time for a given entity (e.g., `Geography` might be treated as static if its characteristics are constant for the forecast horizon).
*   **Dynamic Covariates:** Features that change over time, both known in the future (e.g., `Year`, `Month` if doing sub-annual forecasts) and unknown in the future (e.g., economic indicators).

For `df_population_forecasts_processed`, we could structure the data as follows:

*   **Time Index:** `Date`
*   **Target:** `Value`
*   **Grouping Variables:** `Geography`, `Gender`, `Age` (as each combination defines a unique series)
*   **Static Covariates:** `Geography` (as an embedding)
*   **Dynamic Covariates (known in future):** `Year` (derived from `Date`). Other potential features like `Gender` and `Age` can also be considered dynamic if their influence is modeled as time-varying.

Libraries like `pytorch_forecasting` provide excellent tools to prepare data into the specific `TimeSeriesDataSet` format required by TFTs. This involves defining all the above components and handling features like embeddings for categorical variables and scaling for numerical ones. A full implementation would involve defining the `TimeSeriesDataSet`, then instantiating and training a `TemporalFusionTransformer` model.

<h2 style='color:blue;'>Conceptual Approach for Hybrid Spatiotemporal Models</h2>

Hybrid Spatiotemporal models combine temporal forecasting techniques (like LSTMs or Transformers) with spatial modeling components (e.g., Graph Neural Networks or Convolutional Neural Networks for grid-based data). These models are particularly useful when both 'where' something happens and 'when' it happens are crucial for prediction, and there are dependencies between different locations.

For our urban infrastructure demand forecasting, a hybrid model would be relevant if we wanted to predict demand *across different small areas* (spatial) *over time* (temporal), considering how demand in one area might influence demand in another. Key considerations include:

*   **Spatial Component:** Representing the geographic relationships between different `Geography` entities. This might involve defining a contiguity matrix or a distance-based graph between areas.
*   **Temporal Component:** Using a time-series model (e.g., LSTM, GRU, Transformer) to capture temporal dependencies for each location.
*   **Integration:** How the spatial and temporal components interact. This could be done by stacking layers (e.g., Graph Conv layer followed by LSTM), or more complex architectures with attention mechanisms that integrate spatio-temporal information.

**Data Structure for a Hybrid Model:**

*   **Spatial ID:** A unique identifier for each geographical unit (e.g., `Geography`).
*   **Time Index:** `Date`.
*   **Target Variable:** The variable to be forecasted (e.g., aggregated `Value` for population per `Geography`).
*   **Spatial Features:** Coordinates (`Geo Point` if available and processed), or other geographical attributes.
*   **Time-Series Features:** Aggregated population by `Geography` over time, and any other relevant dynamic covariates.

Implementing such a model typically involves custom model architectures using deep learning frameworks like TensorFlow or PyTorch, defining custom layers for spatial interactions (e.g., Graph Convolutional Layers if a graph structure is used), and carefully designing the network to learn both spatial and temporal dependencies. Given the data's current form, the first step for a hybrid model would be to aggregate the population data by `Geography` and `Date` and then define the spatial relationships between these geographies.

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# Extract the energy consumption values
data = annual_total_energy['Total_Energy_Consumption_kWh'].values.reshape(-1, 1)

# Scale the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Function to create sequences for LSTM
def create_sequences(data, look_back=1):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:(i + look_back), 0])
        y.append(data[i + look_back, 0])
    return np.array(X), np.array(y)

# For this very small dataset, a look_back of 1 is the most feasible.
# A larger look_back would leave too few samples for training.
look_back = 1
X, y = create_sequences(scaled_data, look_back)

# Reshape input to be [samples, time steps, features]
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

print(f"Original data points: {len(data)}")
print(f"Data points after creating sequences (X, y): {len(X)}")
print(f"Shape of X for LSTM: {X.shape}")
print(f"Shape of y for LSTM: {y.shape}")

print("Data prepared for LSTM model.")
```

```python
# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(look_back, 1)))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

model.summary()

# Train the model
# Due to extremely limited data, epochs and batch_size are set low for demonstration.
# In a real-world scenario with more data, these would be significantly higher.
print("\nTraining the LSTM model...")
model.fit(X, y, epochs=50, batch_size=1, verbose=0) # Set verbose to 0 to suppress output per epoch
print("Model training complete.")
```

```python
# Make predictions
# We will use the last available data point to predict the next future step
# Then use that prediction to predict the step after, and so on.

# First, predict on the existing data to see how well it fits
predicted_scaled = model.predict(X)
predicted_original = scaler.inverse_transform(predicted_scaled)

# For forecasting future values, we need to create an input sequence for the next year
# Given our data points are sparse (2011, 2016, 2021, 2026), predicting 2031 (next 5-year interval)
last_value_scaled = scaled_data[-1].reshape(1, look_back, 1)

future_predictions_scaled = []
num_future_steps = 2 # Predict for 2031 and 2036
current_input = last_value_scaled

for _ in range(num_future_steps):
    next_prediction_scaled = model.predict(current_input)
    future_predictions_scaled.append(next_prediction_scaled[0, 0])
    # Update input for the next prediction: the new input sequence includes the latest prediction
    current_input = np.append(current_input[:, 1:, :], next_prediction_scaled.reshape(1, 1, 1), axis=1)
    # If look_back > 1, this needs adjustment to correctly slide the window.
    # For look_back=1, it's just the previous prediction.

future_predictions_original = scaler.inverse_transform(np.array(future_predictions_scaled).reshape(-1, 1))

# Prepare data for plotting
train_years = annual_total_energy['Year'].iloc[look_back:].values
train_actual = annual_total_energy['Total_Energy_Consumption_kWh'].iloc[look_back:].values

# Create a full range of years including future predictions
all_years = list(annual_total_energy['Year'].values)
last_year = all_years[-1]
for i in range(1, num_future_steps + 1):
    all_years.append(last_year + (i * 5)) # Assuming 5-year intervals

# Combine actual and future predictions for plotting
plot_data_actual = annual_total_energy.copy()

future_years = [last_year + (i * 5) for i in range(1, num_future_steps + 1)]
future_df = pd.DataFrame({
    'Year': future_years,
    'Total_Energy_Consumption_kWh': future_predictions_original.flatten()
})

forecast_df = pd.concat([plot_data_actual, future_df], ignore_index=True)

# Visualize the results
plt.figure(figsize=(12, 7))
sns.lineplot(x='Year', y='Total_Energy_Consumption_kWh', data=plot_data_actual, marker='o', label='Actual Data')
sns.scatterplot(x='Year', y='Total_Energy_Consumption_kWh', data=future_df, color='red', s=100, zorder=5, label='LSTM Forecast')
plt.plot(forecast_df['Year'], forecast_df['Total_Energy_Consumption_kWh'], linestyle='--', color='gray', alpha=0.7, label='Forecast Trend')

plt.title('Total Energy Consumption (kWh) Actual vs. LSTM Forecast')
plt.xlabel('Year')
plt.ylabel('Total Energy Consumption (kWh)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(all_years) # Ensure all years including future ones are shown
plt.legend()
plt.tight_layout()
plt.show()

print("\nLSTM Forecasting complete. Please review the plot and consider the data limitations mentioned earlier.")
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
from tqdm.notebook import tqdm
tqdm.pandas()

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="my-geocoder-app")

# Wrap the geocoding function with a rate limiter to avoid overloading the service
# This adds a delay between requests, so it might take some time for large datasets.
geocode = RateLimiter(geolocator.reverse, min_delay_seconds=1)

# Function to extract latitude and longitude and then geocode
def get_location_details(geo_point):
    try:
        # Geo Point is like '-37.83363154012436, 144.98420174077557'
        lat, lon = map(float, geo_point.split(','))
        location = geocode((lat, lon), language='en')
        if location and location.address:
            # Extract a suitable human-readable location, e.g., suburb or city
            # You might need to adjust this based on the Nominatim response structure
            address_parts = location.address.split(', ')
            # Attempt to find a city or suburb name, prioritizing clarity
            # This is a heuristic and might need adjustment based on typical results
            for part in address_parts:
                if 'Melbourne' in part or 'VIC' in part or 'Australia' in part:
                    continue # Skip general city/state/country
                if not part.isdigit() and not any(char.isdigit() for char in part):
                    return part # Return the first non-numeric part as a potential suburb/area
            return address_parts[0] # Fallback to the first part if no specific suburb found
        return None
    except Exception as e:
        # print(f"Error geocoding {geo_point}: {e}")
        return None

print("Extracting location details from Geo Points. This may take a while...")

# Apply the function to the 'Geo Point' column to create a new 'Location' column
df_energy_processed['Location'] = df_energy_processed['Geo Point'].progress_apply(get_location_details)

print("Geocoding complete. Displaying first few rows with new 'Location' column.")
display(df_energy_processed.head())
print(df_energy_processed['Location'].value_counts(dropna=False).head()) # Fixed typo: value_values -> value_counts

# Now, we will recreate df_energy_long using the new 'Location' column in id_vars
```
