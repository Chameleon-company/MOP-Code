```python
import pandas as pd
import matplotlib.pyplot as plt

data = [
    ["Business energy consumption", 60],
    ["Live Music Event Planning", 90],
    ["Foot Traffic Analysis Transport & Safety", 50],
    ["Urban Tree Planting", 50],
    ["Pedestrian and Bus Stops", 40],
    ["Enhancing Pedestrian Safety in Melbourne's Urban Areas", 20],
    ["Argyle Square Bin Fill Level Events", 75],
    ["Smart Bin Optimized Waste Collection Route", 20],
    ["Support Services", 20],
    ["Pedestrian Prediction/Walkability Map of CoM", 30],
    ["Bird Habitat Analysis", 30],
    ["Air Quality Prediction", 50],
    ["Identify potential area for new cafe in Melbourne by using residential data Urban Investor", 90],
    ["Customer Service Analysis", 40],
    ["Art finder â€“ Suicide rates/ number of Physicians", 70],
    ["Business establishments (land use) and jobs", 50],
    ["Impact of greenhouse gas emissions", 45],
    ["Real Estate", 40],
    ["Impact of employment on housing prices", 11],
    ["Tree Canopy Changes", 17]
]

# Create the DataFrame
df = pd.DataFrame(data, columns=['Use-Case', 'Percentage'])

df.head()
```

```python
# Define the bins for the percentage ranges
bins = [0, 24, 49, 74, 100]

# Create a new column 'Range' in the DataFrame that assigns each row to a percentage range bin
df['Range'] = pd.cut(df['Percentage'], bins=bins, labels=['0-24', '25-49', '50-74', '75-100'])

range_counts = df['Range'].value_counts().sort_index()

plt.figure(figsize=(9, 5))
range_counts.plot(kind='bar', color=(0.063, 0.639, 0.498))  # Set the color of the bars to the normalized green color
plt.title('Count of Use-Cases by Percentage Range')
plt.xlabel('% Percentage Range of use cases')
plt.ylabel('Count of Use-Cases')
plt.xticks(rotation=0)  # Keep the x labels horizontal
plt.grid(axis='y', linestyle='--', linewidth=0.7,c='g')
plt.tight_layout()

# Show the plot
plt.show()
```
