## **Required modules**


````python
import requests
import pandas as pd
import numpy as np
from io import StringIO
````

## **Data loading**
Data1 (Pedestrian counting data) loading


````python
# Collect_data function

def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    dataset_id = 'pedestrian-counting-system-monthly-counts-per-hour'
    format = 'csv'


    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit': -1,
        'lang': 'en',
        'timezone': 'UTC',
    }

    response = requests.get(url, params=params)


    if response.status_code == 200:
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f'Request failed with status code {response.status_code}')

dataset_id = 'pedestrian-counting-system-monthly-counts-per-hour'

# Pedestrian counting data
data_1 = collect_data(dataset_id)
data_1.head()
data_1.shape
data_1.info()
````

Data2 (Bus stops for Melbourne Visitor Shuttle) loading

````python
def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    dataset_id = 'bus-stops-for-melbourne-visitor-shuttle'
    format = 'csv'


    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit': -1,
        'lang': 'en',
        'timezone': 'UTC',
    }

    response = requests.get(url, params=params)


    if response.status_code == 200:
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f'Request failed with status code {response.status_code}')

# Bus stop for Melbourne visitor shuttle data
data_2 = collect_data(dataset_id)
data_2.head()
data_2.info()
data_2.shape
````

Data3 (Public artworks, fountains, and monuments) loading

````python
def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    dataset_id = 'public-artworks-fountains-and-monuments'
    format = 'csv'


    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit': -1,
        'lang': 'en',
        'timezone': 'UTC',
    }

    response = requests.get(url, params=params)


    if response.status_code == 200:
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f'Request failed with status code {response.status_code}')

# Public artworks, fountains, and monuments data
data_3 = collect_data(dataset_id)
data_3.head()
data_3.info()
data_3.shape
````
