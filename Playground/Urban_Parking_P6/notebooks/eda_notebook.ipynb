#  Exploratory Data Analysis (EDA)
Urban Parking Demand Project

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("final_cleaned_parking.csv")

df.head()
```

```python
# Dataset info
df.info()

# Missing values
df.isnull().sum()
```

```python
print("Total rows:", df.shape[0])
print("Total parking bays:", df['bay_id'].nunique())
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df['hour'] = df['timestamp'].dt.hour
df['day'] = df['timestamp'].dt.day_name()
```

```python
# Parking occupancy by hour
df.groupby('hour')['occupancy'].mean().plot(kind='line')
plt.title("Parking Occupancy by Hour")
plt.xlabel("Hour")
plt.ylabel("Average Occupancy")
plt.show()
```

```python
# Parking demand by day
df.groupby('day')['occupancy'].mean().plot(kind='bar')
plt.title("Parking Demand by Day")
plt.xlabel("Day")
plt.ylabel("Average Occupancy")
plt.show()
```

```python
# Top busiest parking bays
top_bays = df.groupby('bay_id')['occupancy'].mean().sort_values(ascending=False).head(10)
top_bays.plot(kind='bar')
plt.title("Top 10 Busiest Parking Bays")
plt.xlabel("Bay ID")
plt.ylabel("Average Occupancy")
plt.show()
```

##  Insights


- Parking demand is higher during certain hours of the day, which shows that people park more during busy times like work hours.

- Weekdays have more parking usage than weekends, which means people mostly use parking for office, shopping, or daily activities.

- Some parking bays are used more frequently than others, showing that certain locations are more popular or busy.

- Parking demand changes based on both time and location, so it is not the same everywhere or every time.

- The patterns in the data are consistent and predictable, which makes it easier to forecast future parking demand.
