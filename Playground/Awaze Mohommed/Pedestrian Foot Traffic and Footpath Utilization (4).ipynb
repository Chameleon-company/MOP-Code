```python
import numpy as np
import pandas as pd
import requests
import pandas as pd
from io import StringIO

# function to collect data
def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    dataset_id  = dataset_id
    format='csv'

    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit' : -1,
        'lang' : 'en',
        'timezone' : 'UTC'
    }

    response = requests.get(url,params=params)

    if response.status_code ==200:
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f"Request failed with status code: {response.status_code}")
```

```python
# population_smallarea dataset
dataset_id = 'pedestrian-counting-system-monthly-counts-per-hour'
dataset_id_footpath ='footpaths'
df = collect_data(dataset_id)
df_fpt = collect_data(dataset_id_footpath)
print(len(df))
df.head()
df_fpt.head()
```

```python
sensor_counts = df['sensor_name'].value_counts()
sensors_to_keep = sensor_counts[sensor_counts >= 5000].index.tolist()
df = df[df['sensor_name'].isin(sensors_to_keep)]
```

```python
def remove_outliers(sensor):
    q1 = sensor['total_of_directions'].quantile(0.25)
    q3 = sensor['total_of_directions'].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 2 * iqr
    return sensor[(sensor['total_of_directions'] <= upper_bound)]

df = df.groupby('sensor_name').apply(remove_outliers).reset_index(drop=True)
```

```python
from shapely.geometry import Point
import geopandas as gpd

# Convert pedestrian traffic data to GeoDataFrame
df['geometry'] = df['geo_point_2d'].apply(Point)
df_gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Convert footpaths data to GeoDataFrame
df_fpt['geometry'] = df_fpt['geo_point_2d'].apply(Point)
footpaths_gdf = gpd.GeoDataFrame(df_fpt, geometry='geometry', crs="EPSG:4326")

# Perform spatial join with a small distance tolerance
joined_gdf = gpd.sjoin_nearest(df_gdf, footpaths_gdf, how="left", distance_col="distance", max_distance=0.0001)

# Filter out rows where the distance is greater than the threshold

df_merged = joined_gdf[joined_gdf['distance'] <= 0.0001]

df_merged.head()
```
