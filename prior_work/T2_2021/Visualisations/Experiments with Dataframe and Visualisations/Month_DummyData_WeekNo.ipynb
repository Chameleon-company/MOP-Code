```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
#import dataset from data.csv file
df = pd.read_csv('Dummy_Month.csv')
df.head()
```

```python
del df['Unnamed: 0']
```

```python
df = df.sort_values(["Bay_id", "Date", "Hour","Minutes"], ascending = (True, True, True, True))
df = df.reset_index(drop=True)
df
```

```python
df.loc[df.Date <= '2021-08-10', 'Week_Number'] = '1'
df.loc[(df.Date >= '2021-08-11') & (df.Date <= '2021-08-17'), 'Week_Number'] = '2'
df.loc[(df.Date >= '2021-08-18') & (df.Date <= '2021-08-24'), 'Week_Number'] = '3'
df.loc[(df.Date >= '2021-08-25') & (df.Date <= '2021-08-31'), 'Week_Number'] = '4'
```
