```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import dataset from data.csv file
df = pd.read_csv('Dummy_Data.csv')
df.head()
```

```python
df.loc[df['Bay_id'] == 1001].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(10, 6), rot=0, stacked=True)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Parking Availability at Bay ID: 1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df.loc[df['Bay_id'] == 1001].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(10, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Parking Availability at Bay ID: 1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df.loc[df['Bay_id'] == 1001].groupby('Status').Hour.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=True)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Parking Availability at Bay ID: 1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df.loc[df['Bay_id'] == 1001].groupby('Status').Hour.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Parking Availability at Bay ID: 1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
df.loc[df['Bay_id'] == 1001].groupby('Status').Minutes.value_counts().unstack(0).plot(kind='bar', figsize=(8, 6), rot=0, stacked=False)
plt.xlabel("Minutes", labelpad=14)
plt.ylabel("Count", labelpad=14)
plt.title("Parking Availability at Bay ID: 1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
