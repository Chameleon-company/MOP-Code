```python
import pandas as pd
df1 = pd.read_csv("microclimate_cleaned.csv")

brooklyn_data = df1[df1.Location.eq("Brooklyn")]
brooklyn_data["New Value"] = brooklyn_data["Value"].interpolate()

brighton_data = df1[df1.Location.eq("Brighton")]
brighton_data["New Value"] = brighton_data["Value"].interpolate()

coolaroo_data = df1[df1.Location.eq("Coolaroo")]
coolaroo_data["New Value"] = coolaroo_data["Value"].interpolate()

footscray_data = df1[df1.Location.eq("Footscray")]
footscray_data["New Value"] = footscray_data["Value"].interpolate()

dallas_data = df1[df1.Location.eq("Dallas")]
dallas_data["New Value"] = dallas_data["Value"].interpolate()

boxhill_data = df1[df1.Location.eq("Box Hill")]
boxhill_data["New Value"] = boxhill_data["Value"].interpolate()

macleod_data = df1[df1.Location.eq("Macleod")]
macleod_data["New Value"] = macleod_data["Value"].interpolate()

alphington_data = df1[df1.Location.eq("Alphington")]
alphington_data["New Value"] = alphington_data["Value"].interpolate()

cbd_data = df1[df1.Location.eq("Melbourne CBD")]
cbd_data["New Value"] = cbd_data["Value"].interpolate()

campbellfield_data = df1[df1.Location.eq("Campbellfield")]
campbellfield_data["New Value"] = campbellfield_data["Value"].interpolate()

altonanorth2_data = df1[df1.Location.eq("Altona North 2")]
altonanorth2_data["New Value"] = altonanorth2_data["Value"].interpolate()

altonanorth1_data = df1[df1.Location.eq("Altona North 1")]
altonanorth1_data["New Value"] = altonanorth1_data["Value"].interpolate()

thomastown_data = df1[df1.Location.eq("Thomastown East")]
thomastown_data["New Value"] = thomastown_data["Value"].interpolate()

mri_data = df1[df1.Location.eq("MRI Campbellfield fire")]
mri_data["New Value"] = mri_data["Value"].interpolate()

df2 = pd.concat([brooklyn_data,brighton_data,coolaroo_data,footscray_data,dallas_data,boxhill_data,macleod_data,alphington_data,cbd_data,campbellfield_data,altonanorth2_data,altonanorth1_data,thomastown_data,mri_data])
df2
```

```python
df3 = df2.dropna(subset=['New Value'])
df3
```

```python
df3.to_csv('new.csv',index=False)
```
