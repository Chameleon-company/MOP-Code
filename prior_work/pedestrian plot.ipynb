```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
data=pd.read_csv('stage_2_weekday.csv')
```

```python
point_1=(data.longitude.max()+data.longitude.min())/2
point_2=(data.latitude.max()+data.latitude.min())/2
```

```python
avg_ped=data.iloc[:,:-1].sum(axis=1)
```

```python
top_75=np.percentile(avg_ped,75)
```

```python
downtown_map = folium.Map(location = [point_2,point_1],zoom_start = 15)
```

```python
for i in range(0,len(data)):
    if(avg_ped[i]>top_75):
        c='red'
    else:
        c='green'
    location=[data.latitude[i],data.longitude[i]]
    popup = folium.Popup(max_width=450)
    tooltip=str(data['sensor_description'][i])
    icon=folium.Icon(color=c)
    marker = folium.Marker(location = location, popup = popup, tooltip=tooltip,icon=icon)
    marker.add_to(downtown_map)
```
