```python
from selenium import webdriver

op = webdriver.ChromeOptions()
op.add_argument('headless')

browser = webdriver.Chrome(executable_path='Z:\\Software\\chromedriver', options = op)


browser.get("https://www.realtor.com/international/au/melbourne-city-greater-region-vic/p1")

```
