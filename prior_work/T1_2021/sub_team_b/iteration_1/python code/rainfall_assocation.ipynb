```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('rainfall_dataset.csv')
dayofweek_names = df['day of week'].unique().tolist()
df.head(20)
```

## Day of week

```python
# side by side box plot
plt.figure(figsize=(12,8))
sns.boxplot(x= 'day of week', y= 'Rainfall amount (millimetres)',data =df)
plt.show()
```

```python
dayofweek_names
```

## Overlaid histogram

```python

for dayofweek in dayofweek_names:
    rainfall_vaule = df[df['day of week'] == dayofweek]['Rainfall amount (millimetres)']
    plt.hist(rainfall_vaule, edgecolor = 'black', bins=30)
    plt.title('rainfall histogram on {}'.format(dayofweek))
    plt.xlabel('bins')
    plt.ylabel('counts')
    plt.show()
    plt.clf()
```

## Check stat

```python
from scipy import stats

for i in dayofweek_names:
    print(i,'\n')
    print(stats.describe(df[df['day of week'] == i]['Rainfall amount (millimetres)']))
    print('\n--------------------\n')
```

## Year

```python
year_name = df['Year'].unique().tolist()
df
```

```python
plt.figure(figsize=(12,10))
sns.boxplot(data = df, x = 'Year', y = 'Rainfall amount (millimetres)')
plt.show()
```

```python
for year in year_name:
    rainfall_vaule = df[df['Year'] == year]['Rainfall amount (millimetres)']
    plt.hist(rainfall_vaule, edgecolor = 'black', bins=30)
    plt.title('rainfall histogram on {}'.format(year))
    plt.xlabel('bins')
    plt.ylabel('counts')
    plt.show()
    plt.clf()
```

```python
for i in year_name:
    print(i,'\n')
    print(stats.describe(df[df['Year'] == i]['Rainfall amount (millimetres)']))
    print('\n--------------------\n')
```

## Month

```python
month_name = df['Month'].unique().tolist()
plt.figure(figsize=(12,10))
sns.boxplot(data = df, x = 'Month', y = 'Rainfall amount (millimetres)')
plt.show()
```

```python
for month in month_name:
    rainfall_vaule = df[df['Month'] == month]['Rainfall amount (millimetres)']
    plt.hist(rainfall_vaule, edgecolor = 'black', bins=40)
    plt.title('rainfall histogram on {}'.format(month))
    plt.xlabel('bins')
    plt.ylabel('counts')
    plt.show()
    plt.clf()
```

```python
for month in month_name:
    print(month,'\n')
    print(stats.describe(df[df['Month'] == month]['Rainfall amount (millimetres)']))
    print('\n--------------------\n')
```
