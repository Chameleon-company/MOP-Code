```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('PCS_cleaned.csv')
data1=pd.read_csv('sensor_details.csv')
```

```python
data1
```

```python
data=pd.merge(data, data1, left_on = 'Sensor_ID', right_on='sensor_id', how = 'left', left_index=False)
data=data.drop('sensor_id', axis=1)
data=data.iloc[:,[6,7,8,0,1,2,3,4,5]]
```

```python
data['Date_Time']=pd.to_datetime(data['Date_Time'])
```

```python
data_stage_3 = data[(data['Date_Time']>=dt.datetime(2020,4,1)) & (data['Date_Time']<dt.datetime(2020,6,15))]
data_stage_2 = data[(data['Date_Time']>=dt.datetime(2020,6,15)) & (data['Date_Time']<dt.datetime(2020,7,31))]
data_stage_4 = data[data['Date_Time']>=dt.datetime(2020,7,31)]
```

```python
data_stage_2_weekday = data_stage_2[data_stage_2['Day']<5]
data_stage_3_weekday = data_stage_3[data_stage_3['Day']<5]
data_stage_4_weekday = data_stage_4[data_stage_4['Day']<5]
```

```python
data_stage_2_weekend = data_stage_2[data_stage_2['Day']>=5]
data_stage_3_weekend = data_stage_3[data_stage_3['Day']>=5]
data_stage_4_weekend = data_stage_4[data_stage_4['Day']>=5]
```

```python
location=[]
lng=[]
lat=[]
stage_2_weekday = pd.DataFrame(columns = np.unique(data_stage_2_weekday['Time']),
                               index = np.unique(data_stage_2_weekday['sensor_description']))
for i in np.unique(data_stage_2_weekday['sensor_description']): 
    location.append(data1[data1['sensor_description']==i]['location'].values[0])
    lng.append(data1[data1['sensor_description']==i]['longitude'].values[0])
    lat.append(data1[data1['sensor_description']==i]['latitude'].values[0])
    for j in np.unique(data_stage_2_weekday['Time']):
        stage_2_weekday.loc[i,j] = data_stage_2_weekday[(data_stage_2_weekday['Time']==j) & (data_stage_2_weekday['sensor_description']==i)]['Hourly_Counts'].mean()
stage_2_weekday['location'] = location        
stage_2_weekday['longitude'] = lng 
stage_2_weekday['latitude'] = lat 
```
