```python
client = Socrata("data.melbourne.vic.gov.au", None)
 
#getting the pedestrian count data from city of melbourne website from 2009-05-01 to 
#2020-10-31
results = client.get("b2ak-trbp", limit=4000000)
 
# Convert to pandas DataFrame
results_df = pd.DataFrame.from_records(results)
```

```python
results_df[['Date','Time']] = results_df['date_time'].str.split('T',expand=True)
```

```python
del results_df['date_time']
```

```python
del results_df['time']
```

```python
cols=list(results_df.columns)
cols = [cols[-1]] + cols[:-1]
cols = [cols[-1]] + cols[:-1]
results_df=results_df[cols]
results_df.tail()
```
