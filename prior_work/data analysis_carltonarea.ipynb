```
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt 
import seaborn as sns
```

```
path = '/content/drive/My Drive/DataSets/project _python/pm2.5.csv'
df_1 = pd.read_csv(path)
```

```
df_1.info()
```

```
df_1['local_time']= pd.to_datetime(df_1['local_time']) 
```

```
df_1['Month'] = df_1['local_time'].dt.month
df_1['Day'] = df_1['local_time'].dt.day
```

```
data_arc1050= df_1[df_1['site_id'].isin(['arc1050'])]
```

```
data_arc1050.Month.unique()
```

```
data_arc1050_nov= data_arc1050[data_arc1050['Month'].isin(['11'])]
```

```
data_arc1050_nov['value'].describe()
```

```
ax = sns.lineplot(x="local_time", y="value", hue="site_id",
                  err_style="bars", ci=68, data=data_arc1050_nov)
```
