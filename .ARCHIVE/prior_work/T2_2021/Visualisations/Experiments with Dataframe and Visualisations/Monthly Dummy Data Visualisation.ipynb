```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#import dataset from data.csv file
df = pd.read_csv('Dummy_Month.csv')
df.head()
```

```python
#The amount of observations (rows) and characteristics(columns) in a dataset
print ('The Total Rows in the dataset = ' + str(df.shape[0]))
print ('The Total Columns in the dataset = ' + str(df.shape[1]))
```

```python
Bay_ID = (df['Bay_id'] == 1001)
Multiple_Bay_ID = (df['Bay_id'].isin([1001, 1003, 1005]))
All_Bay_ID = (df['Bay_id'].isin([1001, 1002, 1003, 1004, 1005]))
Parking_Status = (df['Status'] == 'Present')
Day_Of_Week = (df['DayOfWeek'] == 'Wednesday')
Date = (df['Date'] == '2021-08-04')
Hour = (df['Hour'] == 11)
```

```python
sns.set(font_scale=1.5)
df.loc[All_Bay_ID & Parking_Status].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at ALL BAYS", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Bay_ID & Parking_Status].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at Bay ID:1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Multiple_Bay_ID & Parking_Status].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at Bay ID:1001, 1003 & 1005", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[All_Bay_ID & Parking_Status].groupby('Status').Hour.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at ALL BAYS", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Bay_ID & Parking_Status & Day_Of_Week].groupby('Status').Hour.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability on Wednesday at Bay ID:1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[All_Bay_ID & Parking_Status & Day_Of_Week & Hour].groupby('Status').Minutes.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability on Wednesday at 11am at ALL BAYS", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Bay_ID & Parking_Status & Day_Of_Week & Date].groupby('Status').Hour.value_counts().unstack(0).plot(kind='bar', figsize=(12, 6), rot=0, stacked=False)
plt.xlabel("Hour of the Day", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability on Wednesday (4/08) at Bay ID:1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Bay_ID & Parking_Status & Date].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(4, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at Bay ID:1001 on Wednesday (11/08)", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.set(font_scale=1.5)
df.loc[Bay_ID & Parking_Status & Day_Of_Week].groupby('Status').DayOfWeek.value_counts().unstack(0).plot(kind='bar', figsize=(4, 6), rot=0, stacked=False)
plt.xlabel("Day of the Week", labelpad=14)
plt.ylabel("Parking Availabilities", labelpad=14)
plt.title("Parking Availability at Bay ID:1001", y=1)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
