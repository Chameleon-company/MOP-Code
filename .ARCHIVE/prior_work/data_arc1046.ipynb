```
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt 
import seaborn as sns
```

```
path = '/content/drive/My Drive/DataSets/project _python/pm2.5.csv'
df_1 = pd.read_csv(path)
```

```
df_1.info()
```

```
df_1['local_time']= pd.to_datetime(df_1['local_time']) 
```
