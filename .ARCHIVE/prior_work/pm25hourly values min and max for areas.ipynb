
User Story 5: As an Analyst, I need to check the hourly trend of PM2.5 values for in which area is maximum or minimum for all the sensors, so that I can understand the overall concertation of PM 2.5. 





```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import datetime
```

```
data1 = pd.read_csv("Microclimate .csv")
data1.tail(20)
```

```
group_min = data1[data1['Value']<38]
group_min
#group_min['Value'].dropna()

gmin = group_min.groupby(by=['Location','Time']).agg({'Value': ['count']})
gmin.columns = gmin.columns.droplevel(0)

gmin1=pd.DataFrame(data = gmin)
gmin1[:24]
vg_min = gmin1.loc[gmin1.groupby(["Location"])["count"].idxmax()]  
vg_min

fc_min = vg_min.merge(group_min,on=['Location','Time'], how='inner')
fc_min
hourly_min = fc_min.groupby(['Location','Time','count']).agg({'Value':'mean'})
print()
hourly_min
```

```
#group_max = data1.groupby(by=['Location','Time']).agg({"Value": ["max"]}) agg({'Value': ['count']}).agg({'Value': ['max']})
group_max = data1[data1['Value']>38]
group_max
gmax = group_max.groupby(by=['Location','Time']).agg({'Value': ['count']})
gmax.columns = gmax.columns.droplevel(0)

gmax1=pd.DataFrame(data = gmax)
gmax1[:24]
vg = gmax1.loc[gmax1.groupby(["Location"])["count"].idxmax()]  
vg

fc = vg.merge(group_max,on=['Location','Time'], how='inner')
fc
hourly_max = fc.groupby(['Location','Time','count']).agg({'Value':'mean'})
print()
hourly_max
```

```
#min_area = group_mean.groupby(by = ["Location"])['Value mean'].min()
#min_area
hourly_min.to_csv("Min  hourly pm2.5 value of areas.csv")
pd.read_csv("Min  hourly pm2.5 value of areas.csv")
hourly_max.to_csv("Max  hourly pm2.5 value of areas.csv")
pd.read_csv("Max  hourly pm2.5 value of areas.csv")
```
