```python
# Working with completed dwelling
data_C = data2.loc[data2['status'] == 'COMPLETED']
```

```python
data_C = data_C.drop(['Unnamed: 0','status','data_format'], axis =1)
```

```python
data_C.head(2)
```

## Transforming the dataset.
- Need to include a total_dwelling feature, representing the sum of all dwelling
- Need to reshape the dataset to obtain suburds and year as features, and total_dwelling as data
- Change the year_completed in year format

```python
data_C.iloc[:,2:8]
```

```python
# Returning the total dwelling per rows
data_C.iloc[:,2:8].sum(axis=1)
```

```python
# Creating new column nameds Total_Dwelling
data_C['Total_Dwelling'] = data_C.loc[:,['resi_dwellings','studio_dwe','one_bdrm_dwe','two_bdrm_dwe','three_bdrm_dwe','student_apartments']].sum(axis=1)
```

```python
data_C.head(20)
```

```python
# Changing year into Date format
from datetime import datetime
# Checking
date = (pd.to_datetime(data_C['year_completed'],format=('%Y'),exact ='False')).dt.year
date
```

### Obtaining the year.
- changing the column 'year_compelted' into datetime, then used dt.year to extract the year.

```python
data_C['year'] = (pd.to_datetime(data_C['year_completed'],format=('%Y'),exact ='False')).dt.year
```

```python
data_C.head(2)
```

```python
# Need to reshape the dataset for modelling
# 1. selecting features
datashape = data_C.iloc[:,[1,-2,-1]]
datashape.head(2)
```

### Checking the new dataset

## Final format:
- As for the analysis, we only required the total amount of dwelling per year and per area
- Leaving the date as year provides the ability to apply predictive analysis afterwards.

```python
# Grouping the dataset by name and year
shape =datashape.groupby(['clue_small_area','year']).sum()
```
