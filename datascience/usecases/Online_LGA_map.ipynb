### Online LGA map reading code

Tasks for case updates or rewritten tasks. You may find that Load local government areas (LGA) files are used in some cases. <br>

These map files contain .shp etc. Read .shp is not feasible. You need to package a series of files and read them all together. Since there are a series of old cases using local LGA files. Online LGA file reading examples are now provided.

Since City of Melbourne Open Data does not have relevant files, VIC files have been closed for downloading without an account. <br>
The online LGA files use Australian Bureau of Statistics published files. The current version of the document is valid until 2026. <br>

Data Source:

https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files

Administrative coding levels now explained <br>
STE is the state, STE Code of VIC is 2 <br>
SA3 is areas. SA3 Code of City of Melbourn is 20604 <br>

Statistical Area Level 3 definition:

https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/main-structure-and-greater-capital-city-statistical-areas/statistical-area-level-3

```python
import os
import requests
import tempfile
import geopandas as gpd
from zipfile import ZipFile
from io import BytesIO

# URL of the ZIP file
zip_url = 'https://www.abs.gov.au/statistics/standards/australian-statistical-geography-standard-asgs-edition-3/jul2021-jun2026/access-and-downloads/digital-boundary-files/SA2_2021_AUST_SHP_GDA2020.zip'

# Get the contents of a ZIP file
response = requests.get(zip_url)
zip_file_bytes = BytesIO(response.content)

# Create a temporary directory to store the unzipped files
with tempfile.TemporaryDirectory() as temp_dir:
    with ZipFile(zip_file_bytes) as zip_file:
        zip_file.extractall(temp_dir)
    shp_file_name = [f for f in os.listdir(temp_dir) if f.endswith('.shp')][0]
    full_shp_path = os.path.join(temp_dir, shp_file_name)
    LGAgdf = gpd.read_file(full_shp_path)

    # Display
    display(LGAgdf)
```
