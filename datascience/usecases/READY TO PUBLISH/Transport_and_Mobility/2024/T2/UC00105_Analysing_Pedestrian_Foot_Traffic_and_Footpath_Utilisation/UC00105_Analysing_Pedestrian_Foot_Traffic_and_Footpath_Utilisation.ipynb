In this section, we collect the pedestrian traffic data from the City of Melbourne's open datasets. This includes data on pedestrian counts at various sensors across the city.

```python
# function to collect data
def collect_data(dataset_id):
    base_url = 'https://data.melbourne.vic.gov.au/api/explore/v2.1/catalog/datasets/'
    dataset_id  = dataset_id
    format='csv'

    url = f'{base_url}{dataset_id}/exports/{format}'
    params = {
        'select': '*',
        'limit' : -1,
        'lang' : 'en',
        'timezone' : 'UTC'
    }

    response = requests.get(url,params=params)

    if response.status_code ==200:
        url_content = response.content.decode('utf-8')
        dataset = pd.read_csv(StringIO(url_content), delimiter=';')
        return dataset
    else:
        print(f"Request failed with status code: {response.status_code}")
```

 Collecting pedestrian-counting-system-monthly-counts-per-hour dataset and footpaths datasets

## Filtering Sensors with Sufficient Data

```python
sensor_counts = df['sensor_name'].value_counts()
sensors_to_keep = sensor_counts[sensor_counts >= 5000].index.tolist()
df = df[df['sensor_name'].isin(sensors_to_keep)]
```

Remove Outliers

```python
def remove_outliers(sensor):
    q1 = sensor['pedestriancount'].quantile(0.25)
    q3 = sensor['pedestriancount'].quantile(0.75)
    iqr = q3 - q1
    upper_bound = q3 + 2 * iqr
    return sensor[(sensor['pedestriancount'] <= upper_bound)]

df = df.groupby('sensor_name').apply(remove_outliers).reset_index(drop=True)
```

# Data Preprocessing

```python
from shapely.geometry import Point
import geopandas as gpd

# Convert pedestrian traffic data to GeoDataFrame
df['geometry'] = df['geo_point_2d'].apply(Point)
df_gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Convert footpaths data to GeoDataFrame
df_fpt['geometry'] = df_fpt['geo_point_2d'].apply(Point)
footpaths_gdf = gpd.GeoDataFrame(df_fpt, geometry='geometry', crs="EPSG:4326")

# Perform spatial join with a small distance tolerance
joined_gdf = gpd.sjoin_nearest(df_gdf, footpaths_gdf, how="left", distance_col="distance", max_distance=0.0001)

# Filter out rows where the distance is greater than the threshold

df_merged = joined_gdf[joined_gdf['distance'] <= 0.0001]

df_merged.head()
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Step 1: Data Preparation for Time Series
# Assuming df has 'sensing_date' (as datetime) and 'pedestriancount'

# Ensure 'sensing_date' is a datetime type
df['sensing_date'] = pd.to_datetime(df['sensing_date'])

# Group the data by sensing date to get daily pedestrian counts
df_daily = df.groupby('sensing_date')['pedestriancount'].sum().reset_index()

# Set 'sensing_date' as the index
df_daily.set_index('sensing_date', inplace=True)

# Step 2: Visualize the pedestrian counts
plt.figure(figsize=(10,6))
plt.plot(df_daily, label='Daily Pedestrian Count')
plt.title('Daily Pedestrian Counts Over Time')
plt.xlabel('Date')
plt.ylabel('Pedestrian Count')
plt.legend()
plt.grid(True)
plt.show()

# Step 3: Split the data into training and testing sets
train_size = int(len(df_daily) * 0.8)  # 80% for training
train, test = df_daily[:train_size], df_daily[train_size:]

# Step 4: Fit the ARIMA model
# (p, d, q) parameters can be adjusted depending on your data analysis (5, 1, 0) is an example
model = ARIMA(train, order=(6, 1, 1))  # p=5, d=1, q=0 are initial guesses, can be tuned
arima_model = model.fit()

# Step 5: Forecast pedestrian counts for the test set duration
forecast = arima_model.forecast(steps=len(test))

# Step 6: Evaluate the model using Mean Squared Error
mse_arima = mean_squared_error(test, forecast)
print(f"ARIMA Model Mean Squared Error: {mse_arima}")

# Step 7: Plot the actual vs forecasted pedestrian counts
plt.figure(figsize=(12, 6))
plt.plot(train.index, train['pedestriancount'], label='Training Data')
plt.plot(test.index, test['pedestriancount'], label='Actual Test Data', color='blue')
plt.plot(test.index, forecast, label='ARIMA Forecast', color='red')
plt.title('ARIMA Model: Pedestrian Traffic Forecast')
plt.xlabel('Date')
plt.ylabel('Pedestrian Count')
plt.legend()
plt.grid(True)
plt.show()
```
