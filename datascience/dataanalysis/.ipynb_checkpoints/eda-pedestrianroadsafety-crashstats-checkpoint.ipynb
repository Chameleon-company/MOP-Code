# Accidents involving Pedestrians in the City of Melbourne
Pedestrian collisions reported by the Victorian  Police Department from 2008-2020. Each record represents an individual traffic accident event, including the date, time and road location of the accident, road location name, location type, latitude/longitude, event type and victims involved.

**Load in the dataset**

```python
dataset= pd.read_csv('https://raw.githubusercontent.com/Chameleon-company/MOP-Code/master/datascience/usecases/interactive_dependencies/Accidents_Pedestrians_Melbourne_2008to2020.csv')
dataset.head(5)
```

**Learn and explore the data**

```python
# Shape of the Dataset
print("Shape of the dataset : ",dataset.shape)
print()
print('Dataset Info')
print()
# display all columns and their data types
print(dataset.info())
```

```python
#Describing the data
dataset.describe()
```

**Data Cleaning Checks**

We need to make sure the data is appropriately clean before starting analysis work. As a reminder, we should check for:

- Duplicate records
- Consistent formatting
- Missing values
- Obviously wrong values

#### Duplicate Records
How many duplicate transaction records are there? 

```python
dup_rows = dataset.duplicated().sum()
dup_rows 
```

In this case, there is no need to use the drop_duplicates() function because there are no duplicate records.

**Consistent formatting**

Are values of same unit of measure formatted in the same way?
Date values are a common source of inconsistent formatting.

```python
dataset["datetime"] = dataset["DATAccidentDate_accident"]+ ' ' +  dataset["TIMAccidentTime_accident"]
dataset["datetime"]=pd.to_datetime(dataset.datetime) 
```

**Missing Percentage**

How many values are missing within the dataset?

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

newdf = dataset.select_dtypes(include=numerics)
len(newdf.columns)
```

```python
missing_percentage=dataset.isna().sum().sort_values(ascending=False)/len(dataset)
missing_percentage
```

```python
#Plot the percentage of missing values
missing_percentage[missing_percentage !=0].plot(kind='bar')
```

The postcode field originating from the person domain contains the highest rate of missing values at 25%

**Exploring the Data**

```python
# Verify the Number of pedestrian accidents per year 
dataset['year'] = pd.DatetimeIndex(dataset['DATAccidentDate_accident']).year
count_accidents_year = dataset.groupby(['year'])['KEYAccidentNumber'].count().reset_index(name='Count of Accidents per year')
count_accidents_year
```

The number of pedestrian accidents by road location (concatenation of road name + road type) and by accident year (i.e 2008, 2010, 2012). Omiting 2020 as it is not a fully developed accident year because the date range stops at March 2020.

```python
dataset["accident_by_road_loc"] = dataset["CATRoadName_acclocation"]+ ' ' +  dataset["CATRoadType_acclocation"]
dataset_2008_2019=dataset[dataset.datetime.dt.year != 2020]  
count_accidents_loc_year = dataset_2008_2019.groupby(['accident_by_road_loc', 'year'])['KEYAccidentNumber'].count().reset_index(name='Count of Accidents')
count_accidents_loc_year
```

```python
#Plot the count of accidents by location
# The banding in the bar chart represents the distinct locations where accidents occurred
fig = px.bar(count_accidents_loc_year, x="year", y="Count of Accidents", color='year')
fig.update_layout(
    title="Number of accidents by Year (2008-2019)"
    ,xaxis_title="Year"
    ,yaxis_title="Number of accidents"
    
)

fig.show()
```

```python
#Setup days of the week
sunday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 0]
monday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 1]
tuesday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 2]
wednesday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 3]
thursday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 4]
friday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 5]
saturday_time=dataset.datetime[dataset.datetime.dt.dayofweek == 6]                            
```

```python
#Plot each day of the week and show the timeline and count of accident occurrence.
sns.set(style="darkgrid")

fig, axs = plt.subplots(2, 4, figsize=(14, 10))

sns.histplot(data=sunday_time.dt.hour,bins=24, kde=True, color="skyblue", ax=axs[0, 0]).set(title='Sunday', ylim=(0, 35))
sns.histplot(data=monday_time.dt.hour,bins=24, kde=True, color="olive", ax=axs[0, 1]).set(title='Monday', ylim=(0, 35))
sns.histplot(data=tuesday_time.dt.hour,bins=24, kde=True, color="gold", ax=axs[0, 2]).set(title='Tuesday', ylim=(0, 35))
sns.histplot(data=wednesday_time.dt.hour,bins=24, kde=True, color="teal", ax=axs[0,3]).set(title='Wednesday', ylim=(0, 35))
sns.histplot(data=thursday_time.dt.hour,bins=24, kde=True, color="red", ax=axs[1,0]).set(title='Thursday', ylim=(0, 35))
sns.histplot(data=friday_time.dt.hour,bins=24, kde=True, color="green", ax=axs[1, 1]).set(title='Friday', ylim=(0, 35))
sns.histplot(data=saturday_time.dt.hour,bins=24, kde=True, color="black", ax=axs[1,2]).set(title='Saturday', ylim=(0, 35))

plt.show()  
```

**People**
  
**1.   Number of Persons invloved in the accident**

**2.   Number of Persons injured in the accident**

The data frame includes data on how many victims were involved and injured in each pedestrian accident. We can easily represent the percentage of involvement and injuries using a pie plot. 

```python
#Exploring Number of Persons invloved in the accident
import plotly.express as px
dataset['number_of_involved']= dataset['NUMPersonsInvolved_accident']
dataset.loc[dataset['NUMPersonsInvolved_accident'] == 1, 'number_of_involved'] = 'Only one person was invloved'
dataset.loc[dataset['NUMPersonsInvolved_accident'] >= 2 , 'number_of_involved'] = '2-5'
dataset.loc[dataset['NUMPersonsInvolved_accident'] > 5 , 'number_of_involved'] = 'More than 5'
fig = px.pie(dataset,names='number_of_involved', title='Number of Persons invloved in accidents in 2008-2019',color_discrete_sequence=px.colors.diverging.PRGn)
fig.show()
```

```python
#Exploring Number of Persons injured in the accident
import plotly.express as px
dataset['number_of_injured']= dataset['NUMPersonsInjured_accident']
dataset.loc[dataset['NUMPersonsInjured_accident'] == 0, 'number_of_injured'] = 'No person was injured'
dataset.loc[dataset['NUMPersonsInjured_accident'] > 0 , 'number_of_injured'] = 'Persons Injured'

fig = px.pie(dataset,names='number_of_injured', title='Injuries in 2008-2019',color_discrete_sequence=px.colors.diverging.PRGn)
fig.show()
```

```python
#Exploring geographical location of where accidents occurred
sns.scatterplot(x=dataset.NUMLongitude_node,y=dataset.NUMLatitude_node, size=0.001)   
```

```python
m = folium.Map(location=[-37.84442039,144.9785283], control_scale=True, zoom_start=10,attr = "text some",tiles="OpenStreetMap")
dataset_copy=dataset.copy()
for i in range(0,len(dataset_copy)):
  # html to be displayed in the popup 
  html="""
         <h3>Accident Location:</h3> """+\
       """<h5>ROAD_NAME:</h5> """ +str(dataset_copy.iloc[i]['CATRoadName_acclocation'])+\
       """<h5>ROAD_NAME_INT:</h5> """+ str(dataset_copy.iloc[i]['CATRoadNameInt_acclocation'])+\
       """<h5>ROAD_TYPE:</h5> """+ str(dataset_copy.iloc[i]['CATRoadType_acclocation']) +\
       """<h5>ROAD_TYPE_INT:</h5> """+ str(dataset_copy.iloc[i]['CATRoadTypeInt_acclocation'])
   
  iframe = folium.IFrame(html=html, width=200, height=300)
   
  folium.Marker(
  location=[dataset_copy.iloc[i]['NUMLatitude_node'], dataset_copy.iloc[i]['NUMLongitude_node']], 
  popup=folium.Popup(iframe, max_width=2650,parse_html=True) ,
  tooltip=str(dataset_copy.iloc[i]['CATRoadName_acclocation']),
  icon=folium.Icon(color='blue',icon='info-sign'),
  ).add_to(m)
#m.save('accidents.html')
m    
```
