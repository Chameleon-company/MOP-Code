## Use Case 4: Vehicle Classification for Infrastructure Planning

### Problem: Plan infrastructure upgrades and maintenance based on the types and weights of vehicles using the roads.



```python
# Adding a hypothetical weight column based on vehicle class

# Explanation:
# In this use case, the primary goal is to plan infrastructure upgrades and maintenance based on the types and weights of vehicles using the roads.
# However, the dataset provided does not include a weight column, which is essential for assessing the impact of heavy vehicles on road wear and tear.
# Without weight data, it would be impossible to fully address the use case, particularly in terms of evaluating the stress that different vehicles place on the infrastructure.
# To address this gap, I looked up average weights for different types of vehicles based on industry standards and general vehicle classifications.
# These weights are measured in tons and represent the typical weight of each vehicle class as follows:
# - Vehicle class 1: 1.5 tons (Small vehicles like sedans, motorcycles, bicycles, etc.)
# - Vehicle class 2: 2.5 tons (Small towing groups like trailers, caravans)
# - Vehicle class 3: 5.0 tons (Two Axle Truck or Bus)
# - Vehicle class 4: 10.0 tons (Three Axle Truck or Bus)
# - Vehicle class 5: 15.0 tons (Four Axle Truck)
# - Vehicle class 6: 20.0 tons (Three Axle Articulated, or Rigid vehicle and trailer)
# - Vehicle class 7: 25.0 tons (Four Axle Articulated, or Rigid vehicle and trailer)
# - Vehicle class 8: 30.0 tons (Five Axle Articulated, or Rigid vehicle and trailer)
# - Vehicle class 9: 35.0 tons (Six Axle Articulated, or Rigid vehicle and trailer)
# - Vehicle class 10: 40.0 tons (B Double, or Heavy truck and trailer)
# - Vehicle class 11: 50.0 tons (Double Road Train, or Medium articulated vehicle and one dog trailer)
# - Vehicle class 12: 60.0 tons (Triple Road Train, or Heavy truck and three trailers)
# - Vehicle class 13: 5.0 tons (An average estimate for unknown vehicle classes)
# - Motorcycle: 0.2 tons (Motorcycles)
# - Bicycle: 0.015 tons (Bicycles)

# By assigning these weights to each vehicle class, including motorcycles and bicycles, we can approximate the total weight that different roads and segments must bear.
# This allows for a more accurate analysis of the potential wear and tear on infrastructure, and ultimately helps in making informed decisions for road maintenance and upgrades.

# Defining the weight mapping for each vehicle class, including motorcycles and bicycles
weights = {
    'vehicle_class_1': 1.5,  # Small vehicles, motorcycles, etc.
    'vehicle_class_2': 2.5,  # Small towing groups
    'vehicle_class_3': 5.0,  # Two Axle Truck or Bus
    'vehicle_class_4': 10.0, # Three Axle Truck or Bus
    'vehicle_class_5': 15.0, # Four Axle Truck
    'vehicle_class_6': 20.0, # Three Axle Articulated
    'vehicle_class_7': 25.0, # Four Axle Articulated
    'vehicle_class_8': 30.0, # Five Axle Articulated
    'vehicle_class_9': 35.0, # Six Axle Articulated
    'vehicle_class_10': 40.0, # B Double
    'vehicle_class_11': 50.0, # Double Road Train
    'vehicle_class_12': 60.0, # Triple Road Train
    'vehicle_class_13': 5.0,  # Unknown vehicle class (average estimate)
    'motorcycle': 0.2,        # Motorcycles
    'bike': 0.015             # Bicycles
}

# Creating new columns for each vehicle class to represent the total weight for that class on each road segment
# Here, I am multiplying the number of vehicles in each class by their respective weight to get the total weight for that class
for cls, weight in weights.items():
    df[f'{cls}_weight'] = df[cls] * weight
```

```python
df.head()
```

```python
# Finally, adding a new column 'total_weight' that sums up the weight of all vehicle classes for each entry
# This 'total_weight' column will be used in further analysis to assess the overall impact on infrastructure
df['total_weight'] = df[[f'{cls}_weight' for cls in weights.keys()]].sum(axis=1)
```

# ML MODELS 

```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error, r2_score
import pandas as pd
import numpy as np


# List of vehicle class columns from your dataset
vehicle_classes = [
    'vehicle_class_1', 'vehicle_class_2', 'vehicle_class_3', 'vehicle_class_4', 
    'vehicle_class_5', 'vehicle_class_6', 'vehicle_class_7', 'vehicle_class_8', 
    'vehicle_class_9', 'vehicle_class_10', 'vehicle_class_11', 'vehicle_class_12', 
    'vehicle_class_13'
]

# ===========================
# Data Preparation
# ===========================

# Preparing the data (features and targets)
X = df[['speed_limit', 'average_speed', '85th_percentile_speed', 'maximum_speed', 'suburb', 'road_segment'] + vehicle_classes]

# Setting a threshold for high-risk road segments based on total weight
threshold_weight = df['total_weight'].quantile(0.75)

# Creating a binary target variable for classification: 1 if total weight exceeds the threshold, else 0
df['high_risk'] = (df['total_weight'] > threshold_weight).astype(int)

# Features and target for classification
y_classification = df['high_risk']

# Encoding categorical variables (suburb and road_segment) for ML
X = pd.get_dummies(X, columns=['suburb', 'road_segment'], drop_first=True)

# Splitting the data into training and testing sets for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(
    X, y_classification, test_size=0.3, random_state=42
)

# Scaling the data (important for algorithms like RandomForest)
scaler_class = StandardScaler()
X_train_class_scaled = pd.DataFrame(
    scaler_class.fit_transform(X_train_class),
    columns=X_train_class.columns,
    index=X_train_class.index
)
X_test_class_scaled = pd.DataFrame(
    scaler_class.transform(X_test_class),
    columns=X_test_class.columns,
    index=X_test_class.index
)

# ===========================
# Classification Model - Random Forest
# ===========================

# Training the classification model - Random Forest
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_class_scaled, y_train_class)

# Making predictions for classification
y_pred_class = clf.predict(X_test_class_scaled)

# Evaluating the classification model
classification_accuracy = accuracy_score(y_test_class, y_pred_class)
print(f"Classification Accuracy: {classification_accuracy:.2f}")

# ===========================
# Regression Model for Total Weight
# ===========================

# Target variable for predictive modeling (Total Weight as the infrastructure impact indicator)
y_regression = df['total_weight']

# Splitting the data into training and testing sets for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X, y_regression, test_size=0.3, random_state=42
)

# Scaling the data
scaler_reg = StandardScaler()
X_train_reg_scaled = pd.DataFrame(
    scaler_reg.fit_transform(X_train_reg),
    columns=X_train_reg.columns,
    index=X_train_reg.index
)
X_test_reg_scaled = pd.DataFrame(
    scaler_reg.transform(X_test_reg),
    columns=X_test_reg.columns,
    index=X_test_reg.index
)

# Training the regression model - Random Forest
reg = RandomForestRegressor(random_state=42)
reg.fit(X_train_reg_scaled, y_train_reg)

# Making predictions for regression
y_pred_reg = reg.predict(X_test_reg_scaled)

# Evaluating the regression model
mse = mean_squared_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")

# Adding predicted weights back to the test set to analyze
X_test_reg_df = X_test_reg_scaled.copy()
X_test_reg_df['predicted_weight'] = y_pred_reg
X_test_reg_df['actual_weight'] = y_test_reg

# Analyzing top 10 road segments by predicted weight
top_road_segments = X_test_reg_df.nlargest(10, 'predicted_weight')
top_indices = top_road_segments.index

# ===========================
# Predicting Vehicle Class Weights
# ===========================

# Initialize a dictionary to store class-wise predictions
vehicle_class_weight_impact = {}

for vehicle_class in vehicle_classes:
    weight_col = f'{vehicle_class}_weight'
    
    # Check if the weight column exists
    if weight_col not in df.columns:
        print(f"Column {weight_col} does not exist in the DataFrame.")
        continue  # Skip to the next vehicle class
    
    # Define the target variable for this vehicle class
    y_vehicle_class = df[weight_col]
    
    # Use the same train-test split indices
    y_train_vc = y_vehicle_class.loc[X_train_reg.index]
    y_test_vc = y_vehicle_class.loc[X_test_reg.index]
    
    # Training the regression model for the vehicle class
    reg_vc = RandomForestRegressor(random_state=42)
    reg_vc.fit(X_train_reg_scaled, y_train_vc)
    
    # Making predictions on the test set
    y_pred_vc = reg_vc.predict(X_test_reg_scaled)
    
    # Get the predictions and actual values for the top road segments
    y_pred_vc_top = pd.Series(y_pred_vc, index=X_test_reg.index).loc[top_indices]
    y_test_vc_top = y_test_vc.loc[top_indices]
    
    # Store the results
    vehicle_class_weight_impact[vehicle_class] = {
        'predicted_weight': y_pred_vc_top.sum(),
        'actual_weight': y_test_vc_top.sum()
    }

# ===========================
# Displaying the Results
# ===========================

print("Top 10 Road Segments by Predicted Weight Impact:")
print(top_road_segments[['predicted_weight', 'actual_weight']])

print("\nBreakdown of Predicted vs Actual Weights by Vehicle Class in Top Road Segments:")
for vehicle_class, weights in vehicle_class_weight_impact.items():
    print(f"{vehicle_class}: Predicted Total Weight Impact = {weights['predicted_weight']:.2f} tons, Actual Weight = {weights['actual_weight']:.2f} tons")
```
