```python
# Function to extract coordinates with error handling
def extract_coords_safe(coord_string):
    try:
        coords = json.loads(coord_string)
        # Navigate through the nested structure safely
        nested_coords = coords['coordinates']
        # Flatten all levels of nested lists to reach the coordinate pair
        while isinstance(nested_coords[0], list):
            nested_coords = nested_coords[0]
        longitude, latitude = nested_coords
        return longitude, latitude
    except (TypeError, ValueError, IndexError) as e:
        # In case of any error, return NaN values
        return pd.NA, pd.NA

# Apply the function and create new columns 'Geo Shape x' and 'Geo Shape y' with error handling
df[['Geo Shape x', 'Geo Shape y']] = df['Geo Shape'].apply(lambda x: pd.Series(extract_coords_safe(x)))

# Display the updated DataFrame with new columns
df.head()
```

```python
# Dropping the 'Geo Shape' column from df
df = df.drop(columns=['Geo Shape'])
df # we see we successfully removed it
```

```python
# Function to split 'Geo Point' into latitude and longitude
def split_geo_point(geo_point):
    try:
        latitude, longitude = map(float, geo_point.split(','))
        return latitude, longitude
    except ValueError:
        return pd.NA, pd.NA

# Apply the function to create new columns for latitude and longitude
df[['Latitude', 'Longitude']] = df['Geo Point'].apply(lambda x: pd.Series(split_geo_point(x)))

# Remove the 'Geo Point' column
df.drop(columns=['Geo Point'], inplace=True)

# Display the updated DataFrame
df.head()
```
