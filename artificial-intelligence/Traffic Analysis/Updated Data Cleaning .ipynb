```python
# lets check how many nan values for each column in the dataset 
df.isna().sum() # onlt one column seg_descr have 17 nan values out of a total of 4177 rows, lets fix it first
```

```python
# visualizing the columns with nan values 
df['seg_descr'] # we see that the df has 17 nan values and all are objects, so we have 3 options 

# 1 - either fill with mode
# 2- drop the 17 columns 
# 3 - encode the column by either using one hot or label encoder and then fill with mean or median 

# since we have a total of 4177 rows, I believe the best option would be to delete all 17 rows with nan values to avoid 
# any biases in data and it wont affect the dataset at all, Plus we get kind of perfect dataset in that way, so I will be removing
# the nan value columns which are 17 in total so we will have now a total rows of 4160 instead of 4177
```

```python
# dropping nan values 
df = df.dropna(subset=['seg_descr'])
```

```python
# now checking if df has any nan values 
df.isna().values.any() # we get false meaning 0 nan values now, lets check further
```

```python
df.isna().sum() # all zero as we can see
```

```python
# Now converting the geo point column into two separate columns as longitude and latitude 

import warnings

# Ignoring all warnings
warnings.filterwarnings('ignore')
# Splitting the 'Geo Point' column into 'latitude' and 'longitude'
df[['latitude', 'longitude']] = df['Geo Point'].str.split(', ', expand=True)

# Converting the 'latitude' and 'longitude' columns to float
df['latitude'] = df['latitude'].astype(float)
df['longitude'] = df['longitude'].astype(float)

# Dropping the original 'Geo Point' column
df.drop(columns=['Geo Point'], inplace=True)

# Displaying the updated DataFrame
df
```
