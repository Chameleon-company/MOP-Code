```python
# Check for unique values in road_name and location fields
unique_road_names = df['road_name'].unique()
unique_locations = df['location'].unique()
unique_suburbs = df['suburb'].unique()

# Display the unique values
print("Unique Road Names:", len(unique_road_names), unique_road_names)
print("Unique Locations:",  len(unique_locations), unique_locations)
print("Unique Suburbs:", len(unique_suburbs), unique_suburbs)
```

```python
# Replace "street" with "Street" (case-insensitive) 
df['road_name'] = df['road_name'].str.replace(r'\bstreet\b', 'Street', case=False, regex=True)

# Replace "St" with "Street"
df['road_name'] = df['road_name'].str.replace(r'\bSt\b', 'Street', regex=True)

# Replace "&" with "and"
df['location'] = df['location'].str.replace('&', 'and')
df['road_name'] = df['road_name'].str.replace('&', 'and')

# Display a sample of the cleaned data
df[['road_name', 'location']].drop_duplicates().head(20)
```

```python
unique_road_names = df['road_name'].unique()
unique_locations = df['location'].unique()
unique_suburbs = df['suburb'].unique()
print("Unique Road Names:", len(unique_road_names))
print("Unique Locations:",  len(unique_locations))
print("Unique Suburbs:", len(unique_suburbs))
```
