```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv(r"C:\Users\USER\Chameleon AI-IoT\MOP-Code\artificial-intelligence\CyberSecurity\datasets\breach_report.csv")  # Update with your file name

df
```

```python
#  Clean Column Names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df
```

```python
df.info()
```

```python
# Handle Missing Values
df.fillna({
    'state': 'Unknown',
    'covered_entity_type': 'Unknown',
    'type_of_breach': 'Unknown',
    'web_description': 'No description provided'
}, inplace=True)

# Fill 'individuals_affected' using assignment
df['individuals_affected'] = df['individuals_affected'].fillna(df['individuals_affected'].median())
df
```

```python
# Parse Dates
df['breach_submission_date'] = pd.to_datetime(df['breach_submission_date'], errors='coerce')
df['breach_year'] = df['breach_submission_date'].dt.year
df['breach_month'] = df['breach_submission_date'].dt.month
df
```

```python
# Define custom severity classification based on actual thresholds
def classify_severity(x):
    if x < 1000:
        return 'Low'
    elif x < 10000:
        return 'Moderate'
    elif x < 100000:
        return 'High'
    else:
        return 'Critical'
# Apply custom classification
df['severity_level'] = df['individuals_affected'].apply(classify_severity)
# Count distribution to check effectiveness
severity_counts = df['severity_level'].value_counts().reindex(['Low', 'Moderate', 'High', 'Critical'])
severity_counts
```

```python
# Label encoding
le = LabelEncoder()
df['type_of_breach_encoded'] = le.fit_transform(df['type_of_breach'])
df['location_of_breached_information_encoded'] = le.fit_transform(df['location_of_breached_information'])
df['covered_entity_type_encoded'] = le.fit_transform(df['covered_entity_type'])
df
```
